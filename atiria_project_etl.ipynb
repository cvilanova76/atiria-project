```python
import os
from google.colab import drive

# 1. Mount Drive at the Top Level (Required for Interactive UI)
try:
    print("üöÄ Requesting Google Drive access...")
    drive.mount('/content/drive', force_remount=True)
    print("‚úÖ Drive mount successful.")
except Exception as e:
    print(f"‚ùå Drive mount failed: {e}")

def verify_atiria_paths():
    print("\nüîç Starting Directory Verification...")

    # Define your private paths (adjust if your folder names are different)
    PROJECT_PATH = '/content/drive/MyDrive/Atiria_Project'
    RAW_DATA_PATH = os.path.join(PROJECT_PATH, 'data/raw')

    # Check if Project Folder exists
    if os.path.exists(PROJECT_PATH):
        print(f"‚úÖ Project folder found: {PROJECT_PATH}")

        # Check if Raw Data exists and list files
        if os.path.exists(RAW_DATA_PATH):
            files = os.listdir(RAW_DATA_PATH)
            pdfs = [f for f in files if f.endswith('.pdf')]
            print(f"‚úÖ Found {len(pdfs)} PDF(s) in the 'data/raw' folder.")
        else:
            print(f"‚ö†Ô∏è Warning: 'data/raw' folder not found. Please create it at: {RAW_DATA_PATH}")
    else:
        print(f"‚ùå Error: Could not find '{PROJECT_PATH}'.")
        print("Please check if your Drive folder name matches exactly (case-sensitive).")

# Run the verification
verify_atiria_paths()
```
